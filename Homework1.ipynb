```python
#1 For each of the regular expressions below, select all the options that would be completely
#recognized by each of those expressions (if any):

import re

elements = ["ABBA", "BBBB", "AI", "BAI"]

elements2 = ["COA", "COOL", "COLA", "CAO"]

def checkPattern(array , pattern):
    for element in array:
        matches = re.match(pattern , element)
        if matches:
            print(element + ' matches the pattern.')
```

```python
#1.1 ) (A|B)BB?[AEIOU]) 
checkPattern(elements,"(A|B)BB?[AEIOU]")
```

```python
# 3.2  Which sentence (b or c) is the most similar to a) according to Dice? Obtain Dice(a,b)
# and Dice(a,c). Consider a word-level comparison, and round the results to two decimal
# places.

sentenceA = "i really like soccer"
sentenceB = "i like basketball"
sentenceC = "soccer really is my favorite sport"

def dice(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    return round(2 * float(intersection) / union , 2)

dic_sA_sB = dice(sentenceA, sentenceB)
print('DICE : Sentence A / Sentence B = ' + str(dice(sentenceA, sentenceB)) )
print('DICE : Sentence A / Sentence C = ' + str(dice(sentenceA, sentenceC)) )
```
